```python

plt.suptitle('Comparing variables against fertility')


sns.relplot(data=swiss, x= "Agriculture", y='Fertility')
plt.title('Agriculture')

sns.relplot(data=swiss, x= "Education", y='Fertility')
plt.title('Education')

sns.relplot(data=swiss, x= "Examination", y='Fertility')
plt.title('Examination')

sns.relplot(data=swiss, x= "Infant.Mortality", y='Fertility')
plt.title('Infant.Mortality')

sns.relplot(data=swiss, x= "Catholic", y='Fertility')
plt.title('Agriculture')
```
