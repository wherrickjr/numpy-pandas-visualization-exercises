```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import seaborn as sns
```

```python
iris = sns.load_dataset('iris')
```

```python
iris
```

```python
sns.histplot(data = iris, x="petal_length")
plt.show() ##this is what the distribution of petal lengths looks like
```

```python
plt.show(sns.relplot(data = iris, x="petal_length",\
                     y = "petal_width", hue = "species"))
## question 2, yes, there is a strong positive correlation
```

```python
plt.show(sns.relplot(data = iris, x="sepal_length", y = \
                     "sepal_width", hue = \
                     "species"))
##3 it would not be reasonable to guess species based on sepal 
#width/length

##4 using petal length/width would give a fairly accurate species
#prediction
```
