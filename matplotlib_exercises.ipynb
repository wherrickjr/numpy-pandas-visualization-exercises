```python
from random import randint
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.array(range(-49, 51))
y = x**2 - x +2
```

```python
plt.plot(x,y)
plt.title('Quadratic Function $y = x^2$')
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.text(1,100, '(0,0)')
plt.scatter(0,0)

plt.show()

#question 1, plot below shows quadratic function with labels and 
# annotated point at origin
```

```python
square_root = np.array(range(0, 51))
f = square_root ** .5

##question 2 is below
```

```python
plt.plot(square_root,f)
plt.title('Quadratic Function $y = \sqrt{x^2}$')
plt.xlabel('$x$')
plt.ylabel('$y$')
```

```python
cubic = np.array(range(-10, 10))
j = cubic**3
```

```python
plt.plot(cubic,j)
plt.title('Cubic Function $y = x^3$')
plt.xlabel('$x$')
plt.ylabel('$y$')
```

```python
exponential = np.array(range(0, 10))
t = 2 ** exponential
```

```python
plt.plot(exponential,t)
plt.title('Exponsential Function $y = 2^x$')
plt.xlabel('$x$')
plt.ylabel('$y$')
```

```python
half = np.array(range(-49, 1))
other = np.array(range(1,51))
rational = half + other
mm = 1 / (rational +1)
```

```python
plt.plot(rational,mm)
plt.title('Rational Function $y = 1/(x + 1)$')
plt.xlabel('$x$')
plt.ylabel('$y$')
```

```python
fig, axs = plt.subplots(2,3) 
fig.suptitle('Vertically stacked subplots')
axs[0,0].plot(x, y)
axs[0,1].plot(exponential, t)
axs[1,0].plot(square_root, f)
axs[1,1].plot(cubic, j)
axs[0,2].plot(rational, mm)
##these are the subplots for question 3
```

```python
plt.plot(x, y, c='blue', label='$x^2$')
plt.plot(exponential, t, c='orange', label='$2^x$')
plt.plot(square_root, f, c='blue', label='$\sqrt{x}$')
plt.plot(cubic, j, c='red', label='$x^3$')
plt.plot(rational, mm, c='black', label='1/(x+1)$')


plt.legend(loc='upper right')
plt.title('all graphs in one')
plt.xlabel('$x$')
plt.xlim(-10, 10)
plt.ylim(-10, 10)

plt.show()

##plots all functions on one graph #4
```
