```python
###part 2

import numpy as np
import pandas as pd
from pydataset import data


# c1 copy the DatFrames:
users = pd.DataFrame({
    'id': [1, 2, 3, 4, 5, 6],
    'name': ['bob', 'joe', 'sally', 'adam', 'jane', 'mike'],
    'role_id': [1, 2, 3, 3, np.nan, np.nan]
})
users

roles = pd.DataFrame({
    'id': [1, 2, 3, 4],
    'name': ['admin', 'author', 'reviewer', 'commenter']
})
roles

#2 right join will make sure all values in id column are filled
roles.merge(users, how = 'right') #right join justs lists
#the table that is in the parenthesis

#3 outer join, puts all ids under one column and name under one column
roles.merge(users,how = 'outer')
```

```python
roles_nfk = roles[['name']]
roles_nfk
users_nfk = users[['name','id']]
```

```python
users_nfk.merge(roles_nfk, how='outer') ##this is what you get when you drop foregin keys
```

```python
mpg = data('mpg')
```

```python
mpg # 234 rows and 11 columns
```

```python
mpg = mpg.rename(columns = {'cty':'city', 'hwy':'highway', 'displ':'display','cyl':'cycle',\
                            'trans':'transmission','drv':'drive'})
```

```python
mpg.describe() #summary statistics
```

```python
mpg.manufacturer.nunique() ##number of unique manufacturers in the mpg DF
```

```python
mpg.model.nunique() ##number of different models
```

```python
mpg['milage_difference'] = mpg.highway - mpg.city ##adding milage difference
```

```python
mpg['average_milage'] = (mpg.highway + mpg.city)/2 ##adding average milwage
```

```python
mpg['is_automatic'] = mpg.transmission.str.contains('auto') == True
mpg['is_automatic'] = mpg['is_automatic'].replace({True: 'automatic', False: 'manual'})
```

```python
mpg
```

```python
mpg.sort_values('is_automatic')
```

```python
# df.groupby(['column_name', 'other_column']).column_name_you_want_to_perform_functions\
#.agg(['list_of_operations','mean'])
mpg.groupby(['is_automatic']).average_milage.agg(['mean'])


##manual gets better mpg
```
