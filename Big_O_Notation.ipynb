```python
from random import randint
import numpy as np
import matplotlib.pyplot as plt
```

```python
constant = np.array(range(-10, 10))
y1 = 0*constant + 1
```

```python
logrithmic = np.array(range(1, 10))
y2 = np.log(logrithmic)
```

```python
linear = np.array(range(-10, 10))
y3 = linear
```

```python
log_scale = np.array(range(1, 10))
y4 = log_scale * (np.log(log_scale))
```

```python
quadratic = np.array(range(-10, 10))
y5 = quadratic ** 2
```

```python
exponential = np.array(range(0, 10))
y6 = 2 ** exponential
```

```python
factorial = np.array(range(1, 10))
y7 = [np.math.factorial(i) for i in factorial]

```

```python
other = np.array(range(0, 10))
y8 = other ** other
```

```python
plt.plot(constant, y1, c='red', label='O(1)')
plt.plot(logrithmic, y2, c='orange', label='O$\log{n}$')
plt.plot(linear, y3, c='yellow', label='$O(n)$')
plt.plot(log_scale, y4, c='green', label='$O(n \log{n}$')
plt.plot(quadratic, y5, c='blue', label='$O(n^2)$')
plt.plot(exponential, y6, c='indigo', label='$O(2^n)$')
plt.plot(factorial, y7, c='violet', label='$O(n!)$')
plt.plot(other, y8, c='pink', label='$O(n^n)$')
plt.xlabel('Elements')
plt.ylabel('Operations')
plt.xlim(-10, 10)
plt.ylim(, 10)


plt.legend(loc='upper right')
plt.title('Big O Notatioin')
plt.xlabel('$x$')


plt.show()
```
